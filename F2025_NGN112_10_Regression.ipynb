```python
import pandas as pd
Students_df=pd.read_csv('/content/student_scores.csv')
print(Students_df.head())
print(Students_df.describe())
Features=['Hours']
X=Students_df[Features]
y=Students_df['Scores']
print('This is X:\n',X)
print('This is y:\n',y)
X=X.to_numpy()
y=y.to_numpy()
print('This is the new X:\n',X)
X=X.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
#print('This is X_test:\n',X_test)
#print(X_test.shape)
#print(X_train.shape)
#print(X.shape)
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
#from sklearn.svm import SVR
#regressor=SVR(kernel='poly',degree=5)
regressor.fit(X_train,y_train)
import numpy as np
#print('The coefficient of the model is:',np.round(regressor.coef_,2))
#print('The intercept value of the model is',np.round(regressor.intercept_,3))
y_predict=regressor.predict(X_test)
print('The actual values are:\n',y_test)
print('The estimated values are:\n',np.round(y_predict,2))
from sklearn.metrics import mean_squared_error, r2_score
MSE=mean_squared_error(y_test,y_predict)
R2=r2_score(y_test,y_predict)
print('\n\n\nThe MSE is:',MSE)
print('The r2 score is:',R2)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(X_test,y_test,label='Actual')
plt.scatter(X_test,y_predict,label='Predicted')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import fetch_california_housing
housing_data=fetch_california_housing()
X=housing_data.data
y=housing_data.target
feature_names=housing_data.feature_names
target_name=housing_data.target_names
print('The features are:\n',feature_names)
print('The target is:\n',target_name)
print(X[0:5])
print('The shape of the data is:',X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=20)
#print('This is X_test:\n',X_test)
#print(X_test.shape)
#print(X_train.shape)
#print(X.shape)
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
#from sklearn.svm import SVR
#regressor=SVR(kernel='poly',degree=2)
regressor.fit(X_train,y_train)
import numpy as np
#print('The coefficient of the model is:',np.round(regressor.coef_,5))
#print('The intercept value of the model is',np.round(regressor.intercept_,3))
y_predict=regressor.predict(X_test)
#print('The actual values are:\n',y_test)
#print('The estimated values are:\n',np.round(y_predict,2))
from sklearn.metrics import mean_squared_error, r2_score
MSE=mean_squared_error(y_test,y_predict)
R2=r2_score(y_test,y_predict)
print('\n\n\nThe MSE is:',MSE)
print('The r2 score is:',R2)
```
